```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.karate_club_graph()
```

```python
print("Node\tDegree")
for v in G:
    print('{}\t{}'.format(v, G.degree(v)))
```

```python
color = ["#1f78b4"] * 34
color[0] = "red"
color[33] = "green"
nx.draw_circular(G, with_labels=True, node_color=color)
```
